```python
# 1 min runtime ish
diffs = {}
for tx in duplicate_trans_ids:
    dfx = all_dfs['trans_details_cleaned'][
            all_dfs['trans_details_cleaned'].transaction_id == tx
            ].fillna(-1)
    x = dfx.iloc[0] == dfx.iloc[1]
    colmask = x[x == False].index
    dfx = dfx[colmask]
    diffs[tx] = dfx
```

```python
#JONB CODE BELOW

# {('aggregate',),
#  ('aggregate', 'amount'),
#  ('aggregate', 'associations'),
#  ('aggregate', 'process_status'),
#  ('associations',),
#  ('process_status',)}

#aggregate - choose bigger value
#Choose aggregate & amount
#Choose aggregate
#process_status & not filed
#whatever association
#process status & not filed

# Get ids of duplicates
duplicate_trans_ids = all_dfs['trans_details_cleaned'][all_dfs['trans_details_cleaned'].duplicated(subset=['transaction_id'])].transaction_id.tolist()

#Get all rows w/ duplicates
dupsdf = all_dfs['trans_details_cleaned'].loc[all_dfs['trans_details_cleaned'].
                                     transaction_id.isin(duplicate_trans_ids)].sort_values(by=['transaction_id'])
#There's only duplicates of 2
set(dupsdf.groupby(['transaction_id']).size().reset_index(name='count')['count'])

df3 = []
for duptrans in duplicate_trans_ids:
    df1 = dupsdf.loc[dupsdf.transaction_id == duptrans]
    
    #Select row w/ 'Complete' process_status
    if df1[df1.process_status == 'Complete'].shape[0] != 0:
        #aggregate - choose bigger value
        df2 = df1[df1.process_status == 'Complete']['aggregate'].fillna(-1)   #rid nulls
        indexOfSelectedRow = df2[df2 == max(df2)].index[0]

    else:
        df2 = df1['aggregate'].fillna(-1)   #rid nulls
        indexOfSelectedRow = df2[df2 == max(df2)].index[0]
        
    #get index values of rows we want
    df3.append(indexOfSelectedRow)
    
#Drop all rows with possible duplicates
all_dfs2 = all_dfs['trans_details_cleaned'].drop(dupsdf.index)
#Append data we want into trans_details_cleaned
all_dfs3 = all_dfs2.append(dupsdf.loc[df3])
   

```
