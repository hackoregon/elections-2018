```python
from fuzzywuzzy import process
import pickle as pk
```

```python
with open("candidate_names.pkl", "rb") as f:
    candidate_names = pk.load(f)
```

```python
matches = {}
for i in candidate_names:
    candidate_names.remove(i)
    if len(candidate_names) > 1:
        matches[i] = process.extract(i, candidate_names, limit=5)
```

```python
count = 0
for k,v in matches.items():
    if v[0][1] >= 87:
        count += 1
        print(k,v[0])
```

```python
for k,v in matches.items():
    if v[0][1] <= 86 and v[0][1] > 80:
        print(k, '-->', v[0])
```
