```python
data = pd.read_csv("../misc_data/candidates.csv", header=None, names=["name", "count"])
```

```python
names = [str(x) for x in data.name.values]
names = [x for x in names if len(x) > 5]
clean_names = []
for name in names:
    split_name = name.split("_")
    clean_parts = []
    for part in split_name:
        if "(" not in part and len(part) > 1:
            clean_parts.append(part)
        if len(clean_parts) > 2:
            clean_parts = [clean_parts[0], clean_parts[-1]]
    clean_names.append(str.join(" ", clean_parts))
```
